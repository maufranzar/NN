```python
import numpy as np
import matplotlib.pyplot as plt

# Define theta for polar plot
theta = np.linspace(0, 2 * np.pi, 360)

# Define the radiation patterns for half-wave and full-wave dipole antennas
# Half-wave dipole radiation pattern (sin^2(theta))
half_wave_pattern = np.abs(np.sin(theta))

# Full-wave dipole radiation pattern (sin(2*theta)^2)
full_wave_pattern = np.abs(np.sin(2 * theta))

# Plotting the radiation patterns
fig, ax = plt.subplots(1, 2, figsize=(14, 6), subplot_kw={'projection': 'polar'})

# Plot half-wave dipole pattern
ax[0].plot(theta, half_wave_pattern, label='Half-Wave Dipole', linewidth=2)
ax[0].set_title('Radiation Pattern: Half-Wave Dipole', fontsize=14)
ax[0].set_rmax(1.0)
ax[0].grid(True)

# Plot full-wave dipole pattern
ax[1].plot(theta, full_wave_pattern, label='Full-Wave Dipole', linewidth=2, color='orange')
ax[1].set_title('Radiation Pattern: Full-Wave Dipole', fontsize=14)
ax[1].set_rmax(1.0)
ax[1].grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```
